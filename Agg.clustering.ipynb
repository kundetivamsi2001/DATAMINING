```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering 
from sklearn.metrics import silhouette_score
import scipy.cluster.hierarchy as shc
from scipy.spatial.distance import squareform,pdist
a=[0.07,0.85,0.66,0.49,0.80]
b=[0.83,0.14,0.89,0.64,0.46]
point=['p1','p2','p3','p4','p5']
data=pd.DataFrame({'point':point,'a':np.round(a,2),'b':np.round(b,2)})
data=data.set_index('point')
data
cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='single')
print(cluster.fit_predict(data))
#visualizing the Scatterplot
plt.figure(figsize=(8,5))
plt.title("Scatter plot of x,y")
plt.xlabel('column a')
plt.ylabel('column b')
plt.scatter(data['a'],data['b'],c='b',marker='o')
dist=pd.DataFrame(squareform(pdist(data[['a','b']]),'euclidean'),columns=data.index.values,index=data.index.values)
dist
#DEndogram
plt.figure(figsize=(12,5))
plt.title("DENDOHGRAM WITH SINGLE LINKAGE")
dend=shc.dendrogram(shc.linkage(data[['a','b']],method='single'),labels=data.index)
```
