```python
import pandas as pd
from apyori import apriori
st_df=pd.read_csv("Market_Basket_Optimisation.csv",header=None)
st_df

#coverting dataframe into list of lists
l=[]
for i in range(1,7501):
    l.append([str(st_df.values[i,j]) for j in range(0,20)])
#applying apriori algorithm
ass_rules=apriori(l,min_support=0.0045,min_confidence=0.2,min_lift=3,min_length=2)
Association_rules=list(ass_rules)
for i in range(0,len(Association_rules)):
    print(Association_rules[i][0])
print("ASSOCIATION RULES ARE:\n")
for item in Association_rules:
    #first index of the inner list
    #contains base item and add item
    pair=item[0]
    items=[x for x in pair]
    print("Rule:"+ items[0]+ "-->"+items[1])
    #second index of innerlist
    print("Support: "+str(item[1]))
    #third index of the list located at 0 th position
    print("Confidence:"+str(item[2][0][2]))
    print("Lift:"+str(item[2][0][3]))
    print("------------------------------------------------------------------")
```
